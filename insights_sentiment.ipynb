```python
# scrap required imformation from html tags using BeautifulSoup and find_all()
```

```python
def getPage(urls):
  result = requests.get(urls)
  return result

results = map(getPage, url)

title = []
Content = []
my_extract = []

for result in results:
  soup = BeautifulSoup(result.text,"html.parser")

  for Title in soup.find_all('h1',class_="post-title"):
    nt = Title.text
    title.append(nt)
  print(title)

  for headline in soup.find_all('div', class_ = "post-text mt-4"):
    new_head = headline.text
    my_extract.append(new_head)
  print(my_extract)
```

```python
data1 = data.drop(['tokenized_CONTENT','clean_text','content'], axis=1)
```

```python
data2 = data1[['TOPIC','text']].copy()
```

```python
# count all positive and negative words from the text column (cleaned content)
```

```python
def count_matches(row):
    words = row['text'].split()
    p_count = 0
    n_count = 0
    neu_count = 0
    for word in words:
      total_w = len(word)
      for key, values in emotion_words.items():  # Iterate over key-value pairs
        if word in values:
          if key == 'positive':
            p_count += 1
          elif key == 'negative':
            n_count += 1
        # else:
        #     neu_count +=1
    return p_count, n_count   # Return the counts
```

```python
data2[['p_count','n_count']] = data2.apply(count_matches,axis=1,result_type = 'expand')
```

```python
data2
```

```python
# Calculate polarity_score
```

```python
data2['Polarity_Score'] = (data2['p_count']- data2['n_count'])/ ((data2['p_count'] + data2['n_count']) + 0.000001)
```

```python
data2['Polarity_Score'].head()
```

```python
def count_word(row):
    return len(row.split())
```

```python
data2['word_count'] = data2['text'].apply(lambda x: count_word(x))
```

```python
data2['word_count'].head(3)
```

```python
# Subjectivity Score = (Positive Score + Negative Score)/ ((Total Words after cleaning) + 0.000001)
```

```python
data2['Subjectivity_Score'] = (data2['p_count']+ data2['n_count'])/ (data2['word_count'] + 0.000001)
```
