```python
import requests
from bs4 import BeautifulSoup
import numpy as np
```

```python
url = "https://news.google.com/rss/search?q=green%20hydrogen&hl=en-IN&gl=IN&ceid=IN:en"
r = requests.get(url)
soup = BeautifulSoup(r.text,"xml")
print(soup.prettify())
```

```python
Title = soup.find_all(("title"))
len(Title)

```

```python
PUB_DATE = soup.find_all('pubDate')
PUB_DATE
```

```python
import pandas as pd

```

```python
df = pd.DataFrame(list(Title), columns = ['Title'])
df.head(5)
```

```python
#df = df.drop(["Title1"], axis = 1)
#df

# df.drop(['A'], axis=1)
```

```python
df['DATE'] = pd.Series(PUB_DATE)

df
df.info()
```

```python
df['column_name'] = df['DATE'].astype(str)
df
```

```python
df['column_name'] = df['column_name'].map(lambda x: x.lstrip('<pubDate>').rstrip('</pubDate>'))
df.head(3)
```

```python
df = df.drop(["DATE"], axis = 1)
df.head(3)
```

```python
df['Date'] = df["column_name"]
df.head(3)
```

```python
df2 = df.iloc[:-1, :]
df2
```

```python
df2 = df2.drop(['column_name'],axis=1)
```
